### Part One: Weather Data

In `./data/weather.dat` you’ll find daily weather data for Morristown, NJ for June 2002.
Write a function that outputs the day number (column one) with the smallest temperature spread (the maximum temperature is the second column, the minimum the third column).

```python
def find_day_with_smallest_spread():
    f = open('./data/weather.dat')
    min1=300 #filler values that are almost guaranteed to get overwritten (I don't think this is great coding technique, but I wasn't sure what other values to start with)
    maxtemp=0
    mintemp=0
    minDay=1
    count=0
    for line in f.readlines():
        if(count>3): #so we don't try to analyze the header
            maxtemp=float(line.split()[1].strip('*')) #how to extract column 2 -- strip and float to accomodate for weird data things
            mintemp=float(line.split()[2].strip('*')) #how to extract column 3
            if(min1>maxtemp-mintemp): #finding the minimum spread
                min1=maxtemp-mintemp
                minDay=float(line.split()[0].strip("*")) #column 1
        count=count+1
        f.close()
    return "Day: ",minDay, " Spread: ", min1
print find_day_with_smallest_spread()
```

```python
f = open('./data/football.dat')
lines = []
for l in f.readlines():
    line = []
    for c in l.split(' '):
        if (c != '') and (c != '-') and not (c.endswith('.')):
            if c.endswith('\n'):
                line.append(c[:-1])
            else:
                line.append(c)
        else:
            continue
    lines.append(line)
f.close()
lines
```
